```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
x=dataset.iloc[:, [3,4]].values
```

```python
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_blobs
```

```python
x, labels = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)
```

```python
# Apply Spectral Clustering
spectral_clustering = SpectralClustering(n_clusters=4, affinity='nearest_neighbors', random_state=42)
cluster_labels = spectral_clustering.fit_predict(x)
```

```python
# Plot the data points
plt.scatter(x[:, 0], x[:, 1], c=cluster_labels, cmap='viridis', s=50, alpha=0.7, label='Data Points')
plt.title('Spectral Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
