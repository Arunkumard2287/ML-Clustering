```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
x=dataset.iloc[:, [3,4]].values
```

```python
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
```

```python
# Generate synthetic data
x, labels = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)
```

```python
# Apply Mean Shift
mean_shift = MeanShift(bandwidth=1.0)
```

```python
cluster_labels = mean_shift.fit_predict(x)
```

```python
# Get cluster centers
cluster_centers = mean_shift.cluster_centers_
```

```python
# Plot the data points and cluster centers
plt.scatter(x[:, 0], x[:, 1], c=cluster_labels, cmap='viridis', s=50, alpha=0.7, label='Data Points')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.title('Mean Shift Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
