```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
x=dataset.iloc[:, [3,4]].values
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
from sklearn.preprocessing import StandardScaler
```

```python
# Generate synthetic data
x, labels = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(x)
```

```python
# Apply BIRCH
birch = Birch(threshold=0.5, n_clusters=4)
cluster_labels = birch.fit_predict(data_scaled)
```

```python
# Plot the data points
plt.scatter(x[:, 0], x[:, 1], c=cluster_labels, cmap='viridis', s=50, alpha=0.7, label='Data Points')
plt.title('BIRCH Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
